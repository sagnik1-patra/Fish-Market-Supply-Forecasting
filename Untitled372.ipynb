```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Fish Market Supply Forecasting"

RESULTS_CSV = os.path.join(BASE_DIR, "results.csv")

ACCURACY_GRAPH = os.path.join(BASE_DIR, "accuracy_graph.png")
PREDICTION_GRAPH = os.path.join(BASE_DIR, "prediction_graph.png")
COMPARISON_GRAPH = os.path.join(BASE_DIR, "comparison_graph.png")
RESULT_GRAPH = os.path.join(BASE_DIR, "result_graph.png")
ERROR_GRAPH = os.path.join(BASE_DIR, "error_graph.png")

# ============================================================
# LOAD DATA
# ============================================================

results = pd.read_csv(RESULTS_CSV)

actual = results["actual"].values
predicted = results["predicted"].values

# ============================================================
# 1 ACCURACY / LOSS GRAPH
# ============================================================

loss = np.abs(actual - predicted)

fig = plt.figure(figsize=(8, 5))
plt.plot(loss)
plt.title("Prediction Loss (Absolute Error)")
plt.xlabel("Time Index")
plt.ylabel("Loss")

plt.tight_layout()
plt.savefig(ACCURACY_GRAPH, dpi=300)
plt.show()
plt.close(fig)

# ============================================================
# 2 PREDICTION GRAPH
# ============================================================

fig = plt.figure(figsize=(8, 5))
plt.plot(actual)
plt.plot(predicted)
plt.title("Actual vs Predicted Fish Supply")
plt.xlabel("Time Index")
plt.ylabel("Supply")
plt.legend(["Actual", "Predicted"])

plt.tight_layout()
plt.savefig(PREDICTION_GRAPH, dpi=300)
plt.show()
plt.close(fig)

# ============================================================
# 3 COMPARISON GRAPH (TRAIN vs TEST)
# ============================================================

split = int(0.5 * len(actual))

fig = plt.figure(figsize=(8, 5))
plt.plot(actual[:split])
plt.plot(predicted[:split])
plt.plot(actual[split:])
plt.plot(predicted[split:])

plt.title("Train vs Test Prediction Comparison")
plt.xlabel("Time Index")
plt.ylabel("Supply")
plt.legend([
    "Train Actual", "Train Predicted",
    "Test Actual", "Test Predicted"
])

plt.tight_layout()
plt.savefig(COMPARISON_GRAPH, dpi=300)
plt.show()
plt.close(fig)

# ============================================================
# 4 RESULT SCATTER GRAPH
# ============================================================

fig = plt.figure(figsize=(6, 6))
plt.scatter(actual, predicted)
plt.title("Actual vs Predicted (Scatter)")
plt.xlabel("Actual Supply")
plt.ylabel("Predicted Supply")

plt.tight_layout()
plt.savefig(RESULT_GRAPH, dpi=300)
plt.show()
plt.close(fig)

# ============================================================
# 5 ERROR / RESIDUAL GRAPH
# ============================================================

errors = actual - predicted

fig = plt.figure(figsize=(8, 5))
plt.plot(errors)
plt.title("Prediction Error (Residuals)")
plt.xlabel("Time Index")
plt.ylabel("Error")

plt.tight_layout()
plt.savefig(ERROR_GRAPH, dpi=300)
plt.show()
plt.close(fig)

# ============================================================
# CONFIRMATION
# ============================================================

print(" GRAPHS SAVED CORRECTLY (HIGH QUALITY)")
print("", ACCURACY_GRAPH)
print("", PREDICTION_GRAPH)
print("", COMPARISON_GRAPH)
print("", RESULT_GRAPH)
print("", ERROR_GRAPH)
```
