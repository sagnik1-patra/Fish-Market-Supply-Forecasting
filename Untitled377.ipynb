```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Fish Market Supply Forecasting"
DATA_PATH = os.path.join(BASE_DIR, "Table_A-5.1.csv")

MODEL_PATH = os.path.join(BASE_DIR, "bis_model.h5")
SCALER_PATH = os.path.join(BASE_DIR, "bis_scaler.pkl")
CONFIG_PATH = os.path.join(BASE_DIR, "bis_config.yaml")
METRICS_PATH = os.path.join(BASE_DIR, "bis_metrics.json")
RESULTS_PATH = os.path.join(BASE_DIR, "bis_results.csv")
PRED_PATH = os.path.join(BASE_DIR, "bis_predictions.csv")

GRAPH_ACC = os.path.join(BASE_DIR, "bis_accuracy_graph.png")
GRAPH_PRED = os.path.join(BASE_DIR, "bis_prediction_graph.png")
GRAPH_COMP = os.path.join(BASE_DIR, "bis_comparison_graph.png")
GRAPH_RES = os.path.join(BASE_DIR, "bis_result_graph.png")
GRAPH_ERR = os.path.join(BASE_DIR, "bis_error_graph.png")

os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()
target_col = df.select_dtypes(include=np.number).columns[0]
data = df[[target_col]].dropna()

# ============================================================
# SCALING
# ============================================================

scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)

with open(SCALER_PATH, "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# SEQUENCE CREATION
# ============================================================

def create_sequences(data, window):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)

# ============================================================
# BAT ALGORITHM INITIAL POPULATION
# ============================================================

def bat_population(n=8):
    bats = []
    for _ in range(n):
        bats.append({
            "window": random.randint(3, 10),
            "units": random.randint(32, 128),
            "lr": random.choice([0.001, 0.0005, 0.0001])
        })
    return bats

# ============================================================
# AIS REFINEMENT
# ============================================================

def ais_refine(best):
    best["window"] = min(max(best["window"] + random.randint(-1, 1), 3), 10)
    best["units"] = min(max(best["units"] + random.randint(-6, 6), 32), 128)
    return best

# ============================================================
# BA + AIS OPTIMIZATION
# ============================================================

def ba_ais_optimize(bats, iters=4):
    best_global = bats[0].copy()
    best_loss = float("inf")

    for _ in range(iters):
        for bat in bats:
            window = bat["window"]
            units = bat["units"]
            lr = bat["lr"]

            X, y = create_sequences(scaled, window)
            split = int(0.8 * len(X))
            X_train, X_test = X[:split], X[split:]
            y_train, y_test = y[:split], y[split:]

            model = Sequential([
                LSTM(units, input_shape=(window, 1)),
                Dense(1)
            ])
            model.compile(optimizer=Adam(lr), loss="mse")
            model.fit(X_train, y_train, epochs=8, batch_size=16, verbose=0)

            loss = model.evaluate(X_test, y_test, verbose=0)

            if loss < best_loss:
                best_loss = loss
                best_global = bat.copy()

        # AIS mutation on best solution
        best_global = ais_refine(best_global)

    return best_global

# ============================================================
# HYBRID BA + AIS EXECUTION
# ============================================================

bat_pop = bat_population()
best_params = ba_ais_optimize(bat_pop)

WINDOW = best_params["window"]
UNITS = best_params["units"]
LR = best_params["lr"]

X, y = create_sequences(scaled, WINDOW)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ============================================================
# FINAL MODEL
# ============================================================

model = Sequential([
    LSTM(UNITS, input_shape=(WINDOW, 1)),
    Dense(1)
])

model.compile(optimizer=Adam(LR), loss="mse")

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=16,
    validation_data=(X_test, y_test),
    callbacks=[EarlyStopping(patience=10, restore_best_weights=True)],
    verbose=1
)

model.save(MODEL_PATH)

# ============================================================
# PREDICTIONS
# ============================================================

train_pred = scaler.inverse_transform(model.predict(X_train))
test_pred = scaler.inverse_transform(model.predict(X_test))
y_train_inv = scaler.inverse_transform(y_train)
y_test_inv = scaler.inverse_transform(y_test)

# ============================================================
# METRICS
# ============================================================

metrics = {
    "model": "Hybrid BA + AIS + LSTM",
    "window": WINDOW,
    "units": UNITS,
    "learning_rate": LR,
    "train_mae": float(mean_absolute_error(y_train_inv, train_pred)),
    "test_mae": float(mean_absolute_error(y_test_inv, test_pred)),
    "train_rmse": float(np.sqrt(mean_squared_error(y_train_inv, train_pred))),
    "test_rmse": float(np.sqrt(mean_squared_error(y_test_inv, test_pred)))
}

with open(METRICS_PATH, "w") as f:
    json.dump(metrics, f, indent=4)

# ============================================================
# SAVE RESULTS & FUTURE PREDICTION
# ============================================================

pd.DataFrame({
    "actual": y_test_inv.flatten(),
    "predicted": test_pred.flatten()
}).to_csv(RESULTS_PATH, index=False)

future = scaler.inverse_transform(
    model.predict(scaled[-WINDOW:].reshape(1, WINDOW, 1))
)

pd.DataFrame({"future_supply": future.flatten()}).to_csv(PRED_PATH, index=False)

# ============================================================
# GRAPHS (SHOW + SAVE CORRECTLY)
# ============================================================

def save_show(path):
    plt.tight_layout()
    plt.savefig(path, dpi=300)
    plt.show()
    plt.close()

plt.figure(figsize=(8,5))
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("BIS Training Loss")
plt.legend(["Train", "Validation"])
save_show(GRAPH_ACC)

plt.figure(figsize=(8,5))
plt.plot(y_test_inv)
plt.plot(test_pred)
plt.title("BIS Prediction")
plt.legend(["Actual", "Predicted"])
save_show(GRAPH_PRED)

plt.figure(figsize=(6,6))
plt.scatter(y_test_inv, test_pred)
plt.title("BIS Result Scatter")
save_show(GRAPH_RES)

plt.figure(figsize=(8,5))
plt.plot(y_test_inv - test_pred)
plt.title("BIS Error Graph")
save_show(GRAPH_ERR)

# ============================================================
# CONFIG YAML
# ============================================================

with open(CONFIG_PATH, "w") as f:
    yaml.dump(best_params, f)

print(" HYBRID BA + AIS (BIS) PIPELINE COMPLETED SUCCESSFULLY")
```
